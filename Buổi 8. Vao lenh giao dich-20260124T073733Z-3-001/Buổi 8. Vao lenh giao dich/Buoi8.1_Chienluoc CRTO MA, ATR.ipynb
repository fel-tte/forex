# Load du lieu len

```python
import pandas as pd
import ccxt
from datetime import datetime, timedelta

# Khởi tạo kết nối đến sàn Binance
exchange = ccxt.binance() # Khong can API Key, Secret Key

### Step 0: Cac tham so ##########################################################
symbol = 'ETHUSDT'
from_date = (datetime.now() - timedelta(days=0)).strftime('%Y-%m-%d')  # Lấy ngày hôm qua
to_date = (datetime.now() + timedelta(days=0)).strftime('%Y-%m-%d')
timeframe = '1m' # 1m
# Định dạng ngày tháng
since = exchange.parse8601(from_date + 'T00:00:00Z')
end = exchange.parse8601(to_date + 'T00:00:00Z')

### Step 1: Lay du lieu ##########################################################
# Lấy dữ liệu thị trường từ sàn Binance
# 1d
ohlcv = exchange.fetch_ohlcv(symbol, timeframe=timeframe, since=since, limit=1000)

# Chuyển dữ liệu thành DataFrame
data = pd.DataFrame(ohlcv, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])

# Chuyen kieu du lieu thanh datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='ms')  # Chieu dai cua data['Timestamp'] la 12

# Doi ten cot
data = data.rename(columns={'Timestamp': 'Datetime'})

data
```

# Viet lai thanh ham

```python
def loaddataBinance_FromTo(symbol, from_date, to_date, timeframe):
	import pandas as pd
	import ccxt
	from datetime import datetime, timedelta

	# Khởi tạo kết nối đến sàn Binance
	exchange = ccxt.binance()

	# Định dạng ngày tháng
	since = exchange.parse8601(from_date + 'T00:00:00Z')
	end = exchange.parse8601(to_date + 'T00:00:00Z')

	# Lấy dữ liệu thị trường từ sàn Binance
	# 1d
	ohlcv = exchange.fetch_ohlcv(symbol, timeframe=timeframe, since=since, limit=1000)

	# Chuyển dữ liệu thành DataFrame
	data = pd.DataFrame(ohlcv, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
	data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='ms')         
	data = data.rename(columns={'Timestamp': 'Datetime'})

	return data
```

# Chien luoc

```python
import pandas as pd
import plotly.graph_objects as go
import redis
import numpy as np
from datetime import datetime, timedelta

# ##############################################Step 0: Các tham số để lấy dữ liệu###############################
symbol = 'ETHUSDT'
from_date = (datetime.now() - timedelta(days=0)).strftime('%Y-%m-%d')  # Lấy ngày hôm qua
to_date = (datetime.now() + timedelta(days=0)).strftime('%Y-%m-%d')
timeframe = '1m' # 1m

# ##############################################Step 1: Lấy dữ liệu##############################################
data = loaddataBinance_FromTo(symbol, from_date, to_date, timeframe)
# print(data)

# ##############################################Step 2: Chiến lược##############################################  
# Đặt thời gian cho MA và ATR
ma_fast_period = 50 # Thời gian MA nhanh trung binh 50 chu ky
ma_slow_period = 200 # Thời gian MA chậm trung bình 200 chu ky
atr_period = 14
atr_threshold = 40 # Xác định ngưỡng biến động thị trường

import talib # Import vao de su dung
# MA7 > 20 and Close > MA7 => Buy
# Nguoc lai => Sell

data['RSI'] = talib.RSI(data['Close'], timeperiod=14)
data['MA14_RSI'] = talib.SMA(data['RSI'], timeperiod=14) # MA cua RSI

# Tính toán MA nhanh và MA chậm bằng TA-Lib
data['MA_Fast'] = talib.SMA(data['Close'], timeperiod=ma_fast_period) # 50
data['MA_Slow'] = talib.SMA(data['Close'], timeperiod=ma_slow_period) # 200

# Sử dụng TA-Lib để tính ATR, trong đó True Range (TR) được tính toán tự động
data['ATR'] = talib.ATR(data['High'], data['Low'], data['Close'], timeperiod=atr_period)

# ##############################################Step 3: Tín hiệu mua bán##############################################
# Xác định tín hiệu mua
data['Buy_Signal'] = (data['MA_Fast'] >= data['MA_Slow']) & (data['ATR'] <= atr_threshold)
# Xác định tín hiệu bán
data['Sell_Signal'] = (data['MA_Fast'] < data['MA_Slow']) & (data['ATR'] > atr_threshold)

data
```

```python
import talib # Import vao de su dung
# MA7 > 20 and Close > MA7 => Buy
# Nguoc lai => Sell

data['MA7'] = talib.SMA(data['Close'], timeperiod=7)
data['MA20'] = talib.SMA(data['Close'], timeperiod=20)
data['Buy_Signal1'] = (data['MA7'] > data['MA20']) & (data['Close'] > data['MA7'])
data['Sell_Signal1'] = (data['MA7'] < data['MA20']) & (data['Close'] < data['MA7'])

data
```
