```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import time

# Thiết lập URL cơ bản
base_url = "http://127.0.0.1:8000/orderHistory"
account = '2654251'

# Xác định khoảng thời gian
start_date = datetime(2023, 8, 1)  # Ngày bắt đầu
end_date = datetime(2024, 2, 29)   # Ngày kết thúc

# Chuẩn bị DataFrame trống để lưu trữ kết quả
df_full = pd.DataFrame()

while start_date < end_date:
    # Đặt ngày kết thúc của khoảng thời gian 7 ngày hoặc ngày kết thúc cuối cùng, tùy thuộc vào cái nào sớm hơn
    period_end_date = min(start_date + timedelta(days=7), end_date)
    
    # Chuẩn bị các tham số cho yêu cầu API
    params = {
        'account': account,
        'startDate': start_date.strftime("%d/%m/%Y"),
        'endDate': period_end_date.strftime("%d/%m/%Y")
    }
    
    # Gửi yêu cầu GET với các tham số đã chỉ định
    response = requests.get(base_url, params=params)
    
    # Kiểm tra xem yêu cầu có thành công không
    if response.status_code == 200:
        data = response.json()
        print(f"Lịch sử đặt hàng cho tài khoản {account} từ {params['startDate']} đến {params['endDate']} là {data}")
        
        # Thêm dữ liệu vào DataFrame
        df_full = pd.concat([df_full, pd.DataFrame(data['data'])], ignore_index=True)
    else:
        print(f"Lỗi kết nối đến API cho khoảng thời gian từ {params['startDate']} đến {params['endDate']}: {response.status_code}")
    
    # Chuyển sang khoảng thời gian 7 ngày tiếp theo
    time.sleep(11)
    start_date = period_end_date

# In ra DataFrame đầy đủ
print(df_full)
```
