```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Tạo dữ liệu dua tren bai toan tren Slide
data = {
    'mau_vang': [1, 1, 1, 0, 0, 1, 1, 0],  # 1 la mau vang, 0 la khac mau vang: Hươu cao cổ và Hổ vằn màu vàng
    'chieu_cao': [200, 200, 200, 90, 90, 6, 6, 4],  # Hươu cao cổ và Hổ vằn có chiều cao là 10
    'dong_vat': ['huou cao co', 'huou cao co', 'huou cao co', 'voi', 'voi', 'ho van', 'ho van', 'khi']  # Nhãn 3 - 2 - 2 - 1 = 8 con
}

# Chuyen tu dict sang DataFrame
# DataFrame là một cấu trúc dữ liệu 2 chiều, tương tự như bảng trong cơ sở dữ liệu hoặc bảng tính trong Excel
df = pd.DataFrame(data)

# Mã hóa nhãn từ văn bản sang số
# Chuyển đổi nhãn văn bản thành số để có thể sử dụng trong mô hình học máy.
# Trong trường hợp này, chúng ta sẽ mã hóa các nhãn động vật thành các số nguyên tương ứng
df['dong_vat'] = df['dong_vat'].map({'huou cao co': 0, 'voi': 1, 'ho van': 2, 'khi': 3})

# Đặt các đặc trưng vào X và nhãn vào y
# X là ma trận đặc trưng, chứa các thuộc tính đầu vào mà mô hình sẽ sử dụng để dự đoán nhãn, doc lap
# y là vector nhãn, chứa các giá trị đầu ra mà mô hình sẽ cố gắng dự đoán, phu thuoc
# Trong trường hợp này, X chứa hai thuộc tính: 'mau_vang' và 'chieu_cao', trong khi y chứa nhãn động vật tương ứng
X = df[['mau_vang', 'chieu_cao']]
y = df['dong_vat']

# Tạo và huấn luyện mô hình cây quyết định: random_state được sử dụng để đảm bảo rằng kết quả của việc huấn luyện mô hình là nhất quán và 
# có thể tái tạo được; nghĩa là mỗi lần bạn chạy mã này với cùng một dữ liệu đầu vào, bạn sẽ nhận được cùng một mô hình cây quyết định
model = DecisionTreeClassifier(random_state=42) # Random_state là một tham số được sử dụng để đảm bảo rằng kết quả của việc huấn luyện mô hình là nhất quán và có thể tái tạo được
model.fit(X, y)

# Vẽ cây quyết định
fig, ax = plt.subplots(figsize=(12, 8))
tree.plot_tree(model, filled=True, feature_names=['mau_vang', 'chieu_cao'], class_names=['huou cao co', 'voi', 'ho van', 'khi'])
plt.show()
```
