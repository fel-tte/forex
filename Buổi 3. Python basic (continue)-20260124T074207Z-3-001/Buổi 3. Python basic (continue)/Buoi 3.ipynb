# Nhập bán kính đường tròn r. Tính và xuất chu vi, diện tích đường tròn tương ứng.
### HD: cv=2*π*r và dt=π*r*r 

```python
import math # Thu vien thu vien math de tinh toan, built-in

print("Mời bạn nhập vào bán kính đường tròn để tính Chu vi và Diện tích")
r = float(input("Nhập bán kính: ")) # 1. Nhập bán kính đường tròn, 2. Chuyển đổi thành số thực (ep kieu)
print("Bán kính là: ", r)

chu_vi = 2 * math.pi * r # Tinh chu vi đường tròn
print("Chu vi la: ", chu_vi) # In ra chu vi
print(f"Chu vi la: {chu_vi}") # In ra chu vi với f-string

dien_tich = math.pi * r * r # Tinh diện tích đường tròn
print("Diện tích là: ", dien_tich) # In ra diện tích
print(f"Diện tích là: {dien_tich}") # In ra diện tích với f-string
```

# Nhap vao ban kinh, tinh chu vi va dien tich hinh tron

```python
# Nhap vao 1 so, gan cho r goi la ban kinh
r = float(input("Nhập bán kính: ")) # Nhap vao ban kinh, ep kieu sang so thuc

print("Bán kính là: ", r) # In ra bán kính

import math # Thu vien thu vien math de tinh toan, built-in: De tinh pi

# Tinh chu vi
# cv = 2 * math.pi * r # Tinh chu vi đường tròn

# Tinh chu vi dua vao ham
def chu_vi(r):
	return 2 * math.pi * r	
cv = chu_vi(r) # Goi ham tinh chu vi
# In ra chu vi
print("Chu vi la: ", cv) # In ra chu vi

def dien_tich(r):
	return math.pi * r * r # Tinh diện tích đường tròn
dt = dien_tich(r) # Goi ham tinh dien tich
# In dien tich
print("Diện tích là: ", dt) # In ra diện tích
```

# Nhap 2 so a va b tu ban phim va su dung ham Cong2So
### Su dung Try...Except de bat loi: Neu nhap khong dung thi in ra "Ban nhap sai kieu", va bat nhap lai

```python
def Cong2So(a, b):
	tong = a + b
	return tong

# Nhap sai kieu thi bat nhap lai
while True:
	try:
		print("Nhap vao so thu 1")
		a = input() # Nhap so thu 1
		print("Nhap vao so thu 2")
		b = input() # Nhap so thu 2

		print("Tong la: ", Cong2So(int(a), int(b))) # Goi ham Cong2So va in ra tong
		break # Neu nhap dung thi thoat khoi vong lap
	except ValueError: # Neu nhap sai kieu thi bat nhap lai
		print("Ban nhap sai kieu, vui long nhap lai") # In ra loi
```

# Viet ham de tinh chuvi va dientich

```python
# Nhập bán kính đường tròn r. Tính và xuất chu vi, diện tích đường tròn tương ứng.HD: cv=2*π*r và dt=π*r*r
# Bài tập ở trên: Chúng ta viết luông tuồng
# Dùng hàm : Tinh ChuVi va DienTich

import math # Thu vien thu vien math de tinh toan, built-in
def TinhChuVi(r):
	chu_vi = 2 * math.pi * r
	return chu_vi

def TinhDienTich(r):
	dien_tich = math.pi * r * r
	return dien_tich

# Nhập bán kính đường tròn r. Tính và xuất chu vi, diện tích đường tròn tương ứng.HD: cv=2*π*r và dt=π*r*r
r = float(input("Nhập bán kính: ")) # Chuyển đổi thành số thực (ep kieu)
print("Bán kính là: ", r)

chu_vi = TinhChuVi(r) # Tinh chu vi đường tròn bang hàm TinhChuVi, va gan vao chu_vi
print("Chu vi la: ", chu_vi)

dien_tich = TinhDienTich(r) # Tinh diện tích đường tròn bang hàm TinhDienTich, va gan vao dien_tich
print("Diện tích là:", dien_tich)
print(f"Dien tich la: {dien_tich}") # In ra diện tích
```

# Tinh ROI

```python
# Tinh ROI
# Nhap vao chi phi dau tu, nhap vao tong doanh thu, nhap vao chi phi
def TinhROI(chi_phi_dau_tu, tong_doanh_thu, chi_phi):
	roi = (tong_doanh_thu - chi_phi) / chi_phi_dau_tu * 100
	return roi

# Nhap vao chi phi dau tu, tong doanh thu, chi phi
chi_phi_dau_tu = float(input("Nhập chi phí đầu tư: "))	
tong_doanh_thu = float(input("Nhập tổng doanh thu: "))
chi_phi = float(input("Nhập chi phí: "))
roi = TinhROI(chi_phi_dau_tu, tong_doanh_thu, chi_phi) # Goi ham TinhROI va gan vao roi
print("ROI là: ", roi) # In ra ROI
```
