# Kiem tra lich su mua ban

```python
import ccxt
from datetime import datetime

# Khởi tạo kết nối với Binance Futures API bằng API Key và Secret Key
exchange = ccxt.binance({
    'apiKey': api_key,  # Your actual API key
    'secret': secret_key,  # Your actual secret key
    'options': {
        'defaultType': 'future'
    }
})

symbol = 'NEARUSDT'  # Adjust the symbol for ccxt format

# Lấy lịch sử giao dịch mua bán
trade_history = exchange.fetch_my_trades(symbol)

print(trade_history)

# In lịch sử giao dịch
for trade in trade_history:
    # Convert timestamp from milliseconds to a datetime object
    trade['timestamp'] = datetime.fromtimestamp(trade['timestamp'] / 1000)
    print(f"ID: {trade['id']}, Cặp giao dịch: {trade['symbol']}, Loại: {trade['side']}, Giá: {trade['price']}, Số lượng: {trade['amount']}, Thời gian: {trade['timestamp']}")
```
