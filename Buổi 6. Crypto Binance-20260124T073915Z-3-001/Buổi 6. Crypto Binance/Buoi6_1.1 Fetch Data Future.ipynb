```python
import ccxt # ccxt hoac binance
import pandas as pd

# Khởi tạo sàn giao dịch Binance
exchange = ccxt.binance()

# Đặt cặp giao dịch (BTC/USDT) và khoảng thời gian (H1 là mỗi giờ)
symbol = 'BTC/USDT'
timeframe = '1m'

# Lấy lịch sử giá
ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=1000) # Ham/ EndPoint fetch_ohlcv

# Chuyển dữ liệu thành DataFrame
data = pd.DataFrame(ohlcv, columns=['Datetime', 'Open', 'High', 'Low', 'Close', 'Volume'])

# Chuyển đổi timestamp sang dạng ngày tháng
data['Datetime'] = pd.to_datetime(data['Datetime'], unit='ms')

data
```

```python
data.to_csv('Buoi3_1.1 Fetch Data Future.csv')
```

# Ve bieu do nen

```python
import plotly.graph_objects as go

# Vẽ biểu đồ candlestick
fig = go.Figure(data=[go.Candlestick(
    x=data['Datetime'],
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close'],
    increasing_line_color='green',  # Màu nến tăng
    decreasing_line_color='red',    # Màu nến giảm
)])

# Thêm tiêu đề và chỉnh sửa giao diện
fig.update_layout(
    title=f'Biểu đồ Candlestick cho {symbol}',
    xaxis_title='Thời gian',
    yaxis_title='Giá',
    xaxis_rangeslider_visible=False  # Bỏ thanh trượt
)

# Hiển thị biểu đồ
fig.show()
```

# Fetch gia Tick

```python
import ccxt

# Tạo đối tượng kết nối với Binance
binance = ccxt.binance()

# Hàm để lấy giá ticker hiện tại
def fetch_ticker(symbol):
    try:
        ticker = binance.fetch_ticker(symbol)
        print(f"Giá hiện tại của {symbol}: {ticker['last']}")
        return ticker
    except Exception as e:
        print(f"Đã xảy ra lỗi khi lấy ticker: {e}")

# Ví dụ lấy ticker của cặp BTC/USDT
symbol = 'BTC/USDT'
fetch_ticker(symbol)
```
