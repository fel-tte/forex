```python
# Hàm để đặt một lệnh mua
import requests
import os
from binance.client import Client

def orderCRTO(data):
    
    # Lấy dữ liệu từ từ điển
    buy_signal = data['Buy_Signal']    
    sell_signal = data['Sell_Signal']
    symbol = data['Symbol']
    insertdate = data['Insertdate']
    
    order = None

    # Khởi tạo client
    client = Client(api_key, api_secret)

    lot_quantity = 5

    order = None

    try:
        if buy_signal == 'True':  # Kiểm tra xem có phải là giá trị Boolean
            # Đặt lệnh mua
            order = client.order_market_buy(symbol=symbol, quantity=lot_quantity)
        elif sell_signal == 'True':
            # Kiểm tra số dư
            balance = client.get_asset_balance(asset="NEAR")
            # Chỉ định số lượng dựa trên số dư
            lot_quantity = min(lot_quantity, float(balance['free']))
            # Đặt lệnh bán
            order = client.order_market_sell(symbol=symbol, quantity=lot_quantity)
    except Exception as e:
        print(f"An error occurred: {e}")

    print(order)
```

```python
import redis
from datetime import datetime, timedelta

def entryCrypto(hash):
    # Xử lý dữ liệu ở đây
    print("Dữ liệu đã được xử lý", datetime.now())

    # Tạo kết nối Redis
    r = redis.Redis(host='localhost', port=6379, db=15)

    # Hash key
    hash_key = hash

    # Đọc dữ liệu từ hash
    data = r.hgetall(hash_key)
    
    if data:
        # Chuyển đổi dữ liệu từ bytes sang chuỗi nếu cần
        data = {key.decode('utf-8'): value.decode('utf-8') for key, value in data.items()}
        # Xử lý dữ liệu
        orderCRTO(data)
        # Xóa hash sau khi xử lý
        r.delete(hash_key)
        print(f"Hash '{hash_key}' đã được xóa khỏi Redis.")
    else:
        print(f"Không tìm thấy dữ liệu cho hash '{hash_key}'.")
```
