```python
# Hàm để đặt một lệnh mua
import requests
import os

def orderChungKhoan(data):
    
    # Lấy dữ liệu từ từ điển
    buy_signal = data['Buy_Signal']    
    sell_signal = data['Sell_Signal']
    symbol = data['Symbol']
    # insertdate = data['Insertdate']
    
    order = None

    try:
        # Trường họp BUY
        if buy_signal == 'True':
            # URL đã dựng API để sử dụng API
            base_url = "http://127.0.0.1:8000/newOrder"

            # Đưa các tham số vào đúng format lệnh mua
            params = {
                "instrumentID": symbol,
                "market": "VN",
                "buySell": "B",
                "orderType": "ATO",
                "price": 0,
                "quantity": 100,
                "account": 2654251,
                "stopOrder": "false",
                "stopPrice": 0,
                "stopStep": 0,
                "lossStep": 0,
                "profitStep": 0,
                "deviceId": "Local Area Connection 2:00-FF-D6-53-51-3D|Ethernet:A4-4C-C8-1F-0A-34|HotspotShield Network Adapter:00-FF-1B-37-F1-B2|Local Area Connection* 1:F8-34-41-52-B6-98|Local Area Connection* 10:FA-34-41-52-B6-97|Ethernet 2:00-09-0F-FE-00-01|Wi-Fi:F8-34-41-52-B6-97|Bluetooth Network Connection:F8-34-41-52-B6-9B",
                "userAgent": "Python/3.11.6(Windows-10-10.0.17763-SP0); ssi-fctrading/2.4.2"
            }

            # Đẩy dữ liệu lên API
            response = requests.get(base_url, params=params)

            # Xử lý kết quả trả về
            if response.status_code == 200:
                data = response.json()["data"]
                print(f"The stock BUY for account 2654251 is {data}")
            else:
                print(f"Error connecting to API: {response.status_code}")
        
        # Trường hợp SELL        
        elif sell_signal == 'True':        
            # URL đã dựng API để sử dụng API
            base_url = "http://127.0.0.1:8000/newOrder"

            # Đưa các tham số vào đúng format lệnh mua
            params = {
                "instrumentID": symbol,
                "market": "VN",
                "buySell": "S",
                "orderType": "ATO",
                "price": 0,
                "quantity": 100,
                "account": 2654251,
                "stopOrder": "false",
                "stopPrice": 0,
                "stopStep": 0,
                "lossStep": 0,
                "profitStep": 0,
                "deviceId": "Local Area Connection 2:00-FF-D6-53-51-3D|Ethernet:A4-4C-C8-1F-0A-34|HotspotShield Network Adapter:00-FF-1B-37-F1-B2|Local Area Connection* 1:F8-34-41-52-B6-98|Local Area Connection* 10:FA-34-41-52-B6-97|Ethernet 2:00-09-0F-FE-00-01|Wi-Fi:F8-34-41-52-B6-97|Bluetooth Network Connection:F8-34-41-52-B6-9B",
                "userAgent": "Python/3.11.6(Windows-10-10.0.17763-SP0); ssi-fctrading/2.4.2"
            }

            # Đẩy dữ liệu lên API
            response = requests.get(base_url, params=params)

            # Xử lý kết quả trả về
            if response.status_code == 200:
                data = response.json()["data"]
                print(f"The stock SELL for account 2654251 is {data}")
            else:
                print(f"Error connecting to API: {response.status_code}")
            
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
import redis
from datetime import datetime, timedelta

def entryChungKhoan(hash):

    # Xử lý dữ liệu ở đây
    print("Dữ liệu đã được xử lý", datetime.now())

    # Tạo kết nối Redis
    r = redis.Redis(host='localhost', port=6379, db=9) # Voi hashkey OG_CK_Arima, Momentum, RSI_K11VCB.VN
    # Hash key
    hash_key = hash

    # Đọc dữ liệu từ hash
    data = r.hgetall(hash_key)
    
    if data:
        # Chuyển đổi dữ liệu từ bytes sang chuỗi nếu cần
        data = {key.decode('utf-8'): value.decode('utf-8') for key, value in data.items()}
        
        # Xử lý dữ liệu
        orderChungKhoan(data)
        # Xóa hash sau khi xử lý
        r.delete(hash_key)
        print(f"Hash '{hash_key}' đã được xóa khỏi Redis.")
    else:
        print(f"Không tìm thấy dữ liệu cho hash '{hash_key}'.")
```

# Schedule

```python
from datetime import datetime, timedelta
import schedule
import time

def schedule_ChungKhoan():
    entryChungKhoan('OG_CK_Arima, Momentum, RSI_K11VCB.VN')

# Lên lịch hàm scan_market để chạy mỗi 15 phút
schedule.every(10).seconds.do(schedule_ChungKhoan)

while True:
    schedule.run_pending()  # Hàm này được gọi liên tục để kiểm tra xem có công việc nào đã đến thời gian cần thực hiện hay không
    time.sleep(1)
```
