# Chay auto PHUT

```python
from datetime import datetime, timedelta
import schedule
import time
import random

# Danh sách các phút cụ thể bạn muốn hàm được chạy
run_minutes = list(range(0, 60, 1)) # 0,1,2,3,4...59 : Cac phut chung ta can chay
# Phut nao da chay 1 lan roi thi khong chay nua

#############################################################################################################
# Thiết lập thời gian chạy cuối cùng để theo dõi - chua chay lan nao
setattr(scan_market, 'last_run', None) # Set scan_market.last_run = None chua chay lan nao

while True: # Luon luon kiem tra, 1 phut co the kiem tra hang ngan lan
    current_time = datetime.now() # Lay thoi gian hien tai 
    current_hour = current_time.hour # Lay gio hien tai
    current_minute = current_time.minute # Lay phut hien
    current_second = current_time.second # Lay giay hien tai

    # Kiểm tra xem phút hiện tại có nằm trong danh sách các phút cần chạy hàm không
    if current_minute in run_minutes and current_second == 0:
        # Kiểm tra xem đã chạy hàm trong phút hiện tại hay chưa
        last_run = getattr(scan_market, 'last_run', None) # Lay phut cuoi cung da chay, xem co chay lan nao trong phut hien tai hay khong
        if last_run is None or last_run != current_minute: # Neu chua chay lan nao, hoac chua chay trong phut hien tai
            scan_market() # Cho chay
            # Lưu lại phút cuối cùng mà hàm đã chạy
            setattr(scan_market, 'last_run', current_minute)
```
