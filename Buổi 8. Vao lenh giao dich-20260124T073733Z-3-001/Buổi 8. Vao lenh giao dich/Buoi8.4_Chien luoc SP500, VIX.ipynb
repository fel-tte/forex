# 1. Mã và thời gian: Du lieu cua SPY (S & P 500)

# Buoc 1: Load du lieu

```python
import pandas as pd
import yfinance as yf

symbols = ['SPY', '^VIX']  # Danh sách các ký hiệu bạn muốn lấy dữ liệu
start_date = "2024-01-01"
end_date = "2024-03-14"
timeframe = '1d'

def load_data(symbols, start_date, end_date):
    data = yf.download(symbols, start=start_date, end=end_date, interval=timeframe)
    
    return data

data = load_data(symbols, start_date, end_date)

data
```

# Buoc 2: Viet ham kiem tra dieu kien mua hay ban

```python
# Định nghĩa ngưỡng cho VIX
vix_threshold_high = 15  # ví dụ, giá trị này cần được tinh chỉnh
vix_threshold_low = 13   # ví dụ, giá trị này cần được tinh chỉnh

# Xác định tín hiệu mua và bán
# Bởi vì dữ liệu được trả về từ yfinance là MultiIndex DataFrame, chúng ta cần xác định đúng cấu trúc để truy cập dữ liệu

# Điều này giả sử rằng khi VIX cao, đó là tín hiệu bán cho SPY và ngược lại
data['Buy_Signal'] = data[('Close', '^VIX')] < vix_threshold_low
data['Sell_Signal'] = data[('Close', '^VIX')] > vix_threshold_high

data
```

# Buoc 3: Ve bieu do

```python
import plotly.graph_objects as go
import pandas as pd

# Trực quan hóa dữ liệu sử dụng Plotly
fig = go.Figure()

# Lấy cột Close
close_data = data['Close']

# Thêm biểu đồ cho SPY
fig.add_trace(go.Scatter(x=close_data.index, y=close_data['SPY'], mode='lines', name='SPY'))

# Thêm biểu đồ cho VIX
fig.add_trace(go.Scatter(x=close_data.index, y=close_data['^VIX'], mode='lines', name='VIX', yaxis="y2"))

# Thêm tín hiệu mua và bán vào biểu đồ
fig.add_trace(go.Scatter(x=close_data[data['Buy_Signal']].index, y=close_data['SPY'][data['Buy_Signal']], mode='markers', marker=dict(color='Green', size=10, symbol='triangle-up'), name='Buy Signal'))
fig.add_trace(go.Scatter(x=close_data[data['Sell_Signal']].index, y=close_data['SPY'][data['Sell_Signal']], mode='markers', marker=dict(color='Red', size=10, symbol='triangle-down'), name='Sell Signal'))

# Cập nhật layout
fig.update_layout(title='SPY vs VIX with Buy & Sell Signals',
                  xaxis_title='Date',
                  yaxis=dict(title='SPY Price'),
                  yaxis2=dict(title='VIX', overlaying='y', side='right'),
                  # template='plotly_dark'
                  )

# Hiển thị biểu đồ
fig.show()
```
