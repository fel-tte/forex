# Su dung nhieu task trong Shedule

```python
import schedule
import time
from datetime import datetime

def job1():
	print("Chạy công việc 1", datetime.now())

def job2():
	print("Chạy công việc 2", datetime.now())

# Lên lịch cho công việc 1 chạy mỗi giây
schedule.every(30).seconds.do(job1)
# Lên lịch cho công việc 2 chạy mỗi phút
schedule.every(1).minute.do(job2)

while True:
	schedule.run_pending() # Kiểm tra và chạy các công việc đã đến lịch
	# time.sleep(1)
```

# Bai tap ve nha:
Thiet ke autotrade chay: 1m chay 1 lan, moi lan chay la quet data Crypto BTCUSDT 1m va in data ra man hinh

```python
def getData(symbol, timeframe):
	import ccxt # ccxt hoac binance
	import pandas as pd

	# Khởi tạo sàn giao dịch Binance
	exchange = ccxt.binance() # Hoặc có thể sử dụng ccxt.coinbase() nếu muốn kết nối với Coinbase
	# exchange = ccxt.coinbase()
	# exchange = ccxt.mexc()


	# Đặt cặp giao dịch (BTC/USDT) và khoảng thời gian (H1 là mỗi giờ)
	symbol = 'ETH/USDT'
	timeframe = '1m'

	# Lấy lịch sử giá
	ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=1000) # Ham/ EndPoint fetch_ohlcv
	print(type(ohlcv)) # Kiểm tra kiểu dữ liệu của ohlcv
	# Chuyển đổi dữ liệu thành DataFrame của pandas
	df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
	# Chuyển đổi cột 'timestamp' thành định dạng ngày tháng	
	df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
	return df
```

```python
import schedule
import time

i = 0
def scan_marketCrypto():
	print("Đang quét thị trường Crypto...")
	data = getData('ETH/USDT', '1m')  # Gọi hàm getData để lấy dữ liệu
	print(data)   

# Lên lịch hàm scan_market để chạy mỗi giây
schedule.every(1).minute.do(scan_marketCrypto)

while True:
	schedule.run_pending() # Hàm này được gọi liên tục để kiểm tra xem có công việc nào đã đến thời gian cần thực hiện hay không
	# time.sleep(1)
```
