# Vi du 1

```python
import random # Built-in

# Định nghĩa số lần gieo đồng xu 10,000
num_throws = 10000000
# Định nghĩa kết quả có thể có: 'S' cho Sấp, 'N' cho Ngửa
possible_outcomes = ['S', 'N'] # Danh sách chứa các kết quả có thể của việc gieo đồng xu
# Sử dụng một dictionary để đếm số lần xuất hiện của mỗi kết quả
outcome_counts = {'S': 0, 'N': 0}

###################################################################################################
# Thực hiện việc gieo đồng xu
for i in range(1, num_throws + 1, 1): # Chay 1, 2, 3, 4, ....10000 => Chay 10.000 lan
    # Moi lan chay, chon ngau nhien 1 trong 2 kq S hoac N
    outcome = random.choice(possible_outcomes) # S hoac N Ham chon ngau nhien 1 trong 2 ket qua

    # Neu kq la S thi tang so lan xuat hien cua S len 1
    if outcome == 'S':
        outcome_counts['S'] += 1
    # Neu kq la N thi tang so lan xuat hien cua N len 1
    elif outcome == 'N':
        outcome_counts['N'] += 1

outcome_counts
```

```python
# In kết quả
print(f"Số lần Sấp: {outcome_counts['S']}")
print(f"Số lần Ngửa: {outcome_counts['N']}")
```

# Neu chạy 1 tỷ lần thì sao tối ưu

# Built-in ; Tu xay dung trong chuong trinh ; Third Party

```python
import numpy as np # Cai dat numpy

# Định nghĩa số lần gieo đồng xu
num_throws = 10000000  # 10 trieu lần

# Sử dụng numpy để tạo ra 1 tỷ kết quả gieo đồng xu
# np.random.choice(['S', 'N'], num_throws) sẽ tạo một mảng với 1 tỷ phần tử
# Mỗi phần tử là 'S' hoặc 'N' được chọn ngẫu nhiên
outcomes = np.random.choice(['S', 'N'], num_throws) # Chon ngau nhien 1 trong 2 kq

# Dem so lan xuat hien S va N
outcome_counts = np.unique(outcomes, return_counts=True) # Dem so lan xuat hien cua S va N
outcome_counts
```

```python
# 99.5/1.3
```

```python
# Đếm số lần xuất hiện của mỗi kết quả
s_count = np.count_nonzero(outcomes == 'S')
n_count = np.count_nonzero(outcomes == 'N')

# In kết quả
print(f"Số lần Sấp: {s_count}")
print(f"Số lần Ngửa: {n_count}")
```

# Viet lai vi du 1 => Dung list de dem so lan xuat hien

```python
# Định nghĩa số lần gieo đồng xu
num_throws = 1000000  # 1 tỷ lần

# Tạo một list để lưu trữ kết quả của mỗi lần gieo
outcomes = []

# Gieo đồng xu và lưu kết quả vào list
for _ in range(0, num_throws, 1):
    outcomes.append(random.choice(['S', 'N']))

# Đếm số lần xuất hiện của mỗi kết quả
s_count = outcomes.count('S')
n_count = outcomes.count('N')

# In kết quả
print(f"Số lần Sấp: {s_count}")
print(f"Số lần Ngửa: {n_count}")
```
