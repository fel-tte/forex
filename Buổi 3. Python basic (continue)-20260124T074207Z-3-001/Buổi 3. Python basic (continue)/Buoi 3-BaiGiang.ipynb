# Nhap vao ban kinh r, tinh chu vi va dien tich duong tron
# Viet ham tinh va in ra

```python
# Nhap vao ban kinh r, tinh chu vi va dien tich duong tron
# Viet ham tinh va in ra

# Ham tinh chu vi va dien tich duong tron
def tinh_chu_vi(r):
	chu_vi = 2 * math.pi * r
	return chu_vi

def tinh_dien_tich(r):
	dien_tich = math.pi * r**2
	return dien_tich

r = input("Nhập vào bán kính r: ")
try:
	r = float(r)
	chu_vi = tinh_chu_vi(r)
	print(f"Chu vi của đường tròn là: {chu_vi}")
	dien_tich = tinh_dien_tich(r)
	print(f"Diện tích của đường tròn là: {dien_tich}")
	
except:
	print("Lỗi")
```

# Bai tap: Viet ham lay du lieu gia BTC tu Binance

```python
import ccxt
binance = ccxt.binance()

# Lay du lieu gia BTC tu Binance
def lay_gia_btc(symbol):
    try:
        ticker = binance.fetch_ticker(f'{symbol.upper()}/USDT')
        return ticker
    except Exception as e:
        print(f"Lỗi: {e}")
        return None
```

```python
def loaddataBinance_FromTo(symbol, from_date, to_date, timeframe):
    import pandas as pd
    import ccxt
    from datetime import datetime

    # Khởi tạo kết nối đến sàn Binance
    exchange = ccxt.binance()

    # Định dạng ngày tháng
    since = exchange.parse8601(from_date + 'T00:00:00Z')
    end = exchange.parse8601(to_date + 'T00:00:00Z')

    # Lấy dữ liệu thị trường từ sàn Binance
    # 1d
    ohlcv = exchange.fetch_ohlcv(symbol, timeframe=timeframe, since=since, limit=1000)

    # Chuyển dữ liệu thành DataFrame
    data = pd.DataFrame(ohlcv, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
    data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='ms')         
    data = data.rename(columns={'Timestamp': 'Datetime'})

    return data
```

```python
symbol = input("Nhập vào symbol: ")

ticker = lay_gia_btc(symbol)

print("Gia cua", symbol, "la:", ticker)
print(f"Gia cua {symbol} la: {ticker}")
```
