```python
def getData(symbol, timeframe):
	import ccxt # ccxt hoac binance
	import pandas as pd

	# Khởi tạo sàn giao dịch Binance
	exchange = ccxt.binance() # Hoặc có thể sử dụng ccxt.coinbase() nếu muốn kết nối với Coinbase
	# exchange = ccxt.coinbase()
	# exchange = ccxt.mexc()


	# Đặt cặp giao dịch (BTC/USDT) và khoảng thời gian (H1 là mỗi giờ)
	symbol = 'ETH/USDT'
	timeframe = '1m'

	# Lấy lịch sử giá
	ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=1000) # Ham/ EndPoint fetch_ohlcv
	print(type(ohlcv)) # Kiểm tra kiểu dữ liệu của ohlcv
	# Chuyển đổi dữ liệu thành DataFrame của pandas
	df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
	# Chuyển đổi cột 'timestamp' thành định dạng ngày tháng	
	df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
	return df
```

# Viet lai While True

```python
from datetime import datetime, timedelta
import schedule
import time
import random

# Khai bao phut chay
run_minutes = list(range(0, 60, 2)) # 0,1,2,3,4...59 : Cac phut chung ta can chay

# Thiết lập thời gian chạy cuối cùng để theo dõi - chua chay lan nao
setattr(scan_marketCrypto, 'last_run', None)
while True:
    current_time = datetime.now() # Lay thoi gian hien tai 
    current_hour = current_time.hour # Lay gio hien tai
    current_minute = current_time.minute # Lay phut hien tai cua may tinh (cung voi server)
    current_second = current_time.second # Lay giay hien tai
    
    if current_minute in run_minutes and current_second == 5: # Phut hien tai nam trong danh sach phut chay va giay hien tai la 5
        # Kiểm tra xem đã chạy hàm trong phút hiện tại hay chưa
        last_run = getattr(scan_marketCrypto, 'last_run', None) # Lay phut cuoi cung ma ham da chay
        if last_run is None or last_run != current_minute: # Neu chua chay lan nao, hoac chua chay trong phut hien tai
            scan_marketCrypto() # Cho chay
            # Lưu lại phút cuối cùng mà hàm đã chạy
            setattr(scan_marketCrypto, 'last_run', current_minute)
    time.sleep(1)
```
