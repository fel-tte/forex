# Lay theo period va interval

```python
import yfinance as yf

ticker = yf.Ticker("STB.VN")  # Thay thế "AAPL" bằng mã cổ phiếu mong muốn

# Lấy dữ liệu theo timeframe 5 phút
data = ticker.history(period="1mo", interval="5m")

print(data)
```

# Lay theo start, end va interval

```python
import yfinance as yf

ticker = yf.Ticker("STB.VN")  # Thay thế "AAPL" bằng mã cổ phiếu mong muốn

# Lấy dữ liệu từ ngày 1/1/2020 đến ngày 31/12/2020
data = ticker.history(start="2024-11-01", end="2024-12-06", interval="5m")

print(data)
```

# Ticker.info: Hàm này trả về thông tin chi tiết về cổ phiếu, bao gồm giá hiện tại, giá mở cửa, giá cao nhất, giá thấp nhất, khối lượng giao dịch, v.v.
### Ticker.financials: Hàm này trả về báo cáo tài chính của công ty, bao gồm bảng cân đối kế toán, báo cáo thu nhập, báo cáo lưu chuyển tiền tệ, v.v.
### Ticker.major_holders: Hàm này trả về danh sách các cổ đông lớn của công ty.
### Ticker.institutional_holders: Hàm này trả về danh sách các tổ chức đầu tư nắm giữ cổ phiếu của công ty.

```python
import yfinance as yf

ticker = yf.Ticker("STB.VN")

# Lấy thông tin chi tiết về cổ phiếu
info = ticker.info
print(info)

# Lấy báo cáo tài chính của công ty
financials = ticker.financials
print(financials)

# Lấy danh sách các cổ đông lớn của công ty
major_holders = ticker.major_holders
print(major_holders)

# Lấy danh sách các tổ chức đầu tư nắm giữ cổ phiếu của công ty
institutional_holders = ticker.institutional_holders
print(institutional_holders)
```

# Ticker.actions: Hàm này trả về danh sách các sự kiện quan trọng của công ty, bao gồm chia cổ tức, tách cổ phiếu, v.v.
### Ticker.splits: Hàm này trả về danh sách các sự kiện tách cổ phiếu của công ty.
### Ticker.dividends: Hàm này trả về danh sách các sự kiện chia cổ tức của công ty.
### Ticker.calendars: Hàm này trả về danh sách các sự kiện quan trọng của công ty, bao gồm ngày công bố lợi nhuận, ngày chia cổ tức, v.v.

```python
import yfinance as yf

ticker = yf.Ticker("AAPL")

# Lấy danh sách các sự kiện quan trọng của công ty
actions = ticker.actions
print(actions)

# Lấy danh sách các sự kiện tách cổ phiếu của công ty
splits = ticker.splits
print(splits)

# Lấy danh sách các sự kiện chia cổ tức của công ty
dividends = ticker.dividends
print(dividends)
```
