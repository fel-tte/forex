```python
# Lay tong so lenh
def getTotalPositions():
    import MetaTrader5 as mt5
    # display data on the MetaTrader 5 package
    # print("MetaTrader5 package author: ",mt5.__author__)
    # print("MetaTrader5 package version: ",mt5.__version__)
    
    # establish connection to MetaTrader 5 terminal
    if not mt5.initialize():
        print("initialize() failed, error code =",mt5.last_error())
        quit()
    
    # check the presence of open positions
    positions_total=mt5.positions_total()
    if positions_total>0:
        # print("Total positions=",positions_total)
        return positions_total
    else:
        # print("Positions not found")
        return 0
    
    # shut down connection to the MetaTrader 5 terminal
    mt5.shutdown()
```

```python
import redis
 # Tạo kết nối Redis
r = redis.Redis(host='localhost', port=6379, db=15) # Theo ben OG

# Hash key
hash_key = "OG_FX_Buoi 5"

# Đọc dữ liệu từ hash
data = r.hgetall(hash_key)

if data:
    # Chuyển đổi dữ liệu từ bytes sang chuỗi nếu cần
    data = {key.decode('utf-8'): value.decode('utf-8') for key, value in data.items()}

    # Xử lý dữ liệu vao lenh, duoc hay khong thi cung xoa hash
    # Xử lý dữ liệu vao lenh
    if getTotalPositions() > 3:
        print("So lenh da mo nhieu hon 3: ", getTotalPositions())
    else:
        print("So lenh chua du 4: ", getTotalPositions())
        entryForex(data)

    # Xoa hash sau khi xu ly
    r.delete(hash_key)
    print(f"Hash '{hash_key}' đã được xóa khỏi Redis.")
else:
    print(f"Không tìm thấy dữ liệu cho hash '{hash_key}'.")
```
