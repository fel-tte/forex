# Lay balance

```python
import pandas as pd
import requests # Python goi restful API, link web
account = '2654251'
url = f"http://127.0.0.1:8000/stockAccountBalance?account={account}" # API noi bo de goi len SSI ~ FastAPI ~ Restful API ~ API

response = requests.get(url) # Goi API

data_dict = response.json()['data']

# Kiểm tra cấu trúc dữ liệu trước khi tạo DataFrame
if isinstance(data_dict, dict):
    # Nếu là dict với các giá trị scalar, cần tạo index
    if all(isinstance(v, (str, int, float)) for v in data_dict.values()):
        data = pd.DataFrame([data_dict])  # Tạo DataFrame với 1 hàng
    else:
        data = pd.DataFrame(data_dict)
elif isinstance(data_dict, list):
    # Nếu là list, tạo DataFrame bình thường
    data = pd.DataFrame(data_dict)
else:
    # Nếu là giá trị scalar, tạo DataFrame với index
    data = pd.DataFrame([data_dict])

data
```

# Buoc 3: Giao dich (Buy/ Sell)

```python
import requests

# Base URL for the API

base_url = "http://127.0.0.1:8000/newOrder" # Endpoint ~ Function

# Parameters for the API request ~ Input
params = {
    "instrumentID": "VCB",
    "market": "VN",
    "buySell": "B",
    "orderType": "ATO",
    "price": 0,
    "quantity": 100,
    "account": '2654251', # Doi lai account cua cac ban
    "stopOrder": "false",
    "stopPrice": 0,
    "stopStep": 0,
    "lossStep": 0,
    "profitStep": 0,
    "deviceId": "Ethernet:00-E0-24-4B-60-9F",
    "userAgent": "Python/3.11.6(Windows-10-10.0.19045-SP0); ssi-fctrading/2.5.3"
}

# Sending the GET request to the API
response = requests.get(base_url, params=params)

# Print the raw response
print(response)
print(response.status_code)
print(response.text)
print(response.json())

# Handling the response
if response.status_code == 200:
    data = response.json()["data"]
    print(f"The stock order for account 2654251 is {data}")
else:
    print(f"Error connecting to API: {response.status_code}")
```
