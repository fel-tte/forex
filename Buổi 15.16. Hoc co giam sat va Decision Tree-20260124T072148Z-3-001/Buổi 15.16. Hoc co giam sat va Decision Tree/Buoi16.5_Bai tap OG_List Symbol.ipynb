```python
import ccxt
from datetime import datetime
# Khởi tạo kết nối đến sàn Binance
exchange = ccxt.binance()

# Lay toan bo symbol tren Binance
symbol_list = exchange.load_markets()
symbol_list
```

```python
# Loc lai lay symbol co kieu "USDT" va price > 1000
symbol_list = [symbol for symbol in symbol_list if symbol.endswith('USDT')]
symbol_list
```

```python
from datetime import datetime, timedelta

import sys
sys.path.append("../Common")
import CommonBinanceDWH

print("Bat dau thuc hien:")
print(datetime.now())

symbol_list = ["BTCUSDT"]
from_date = (datetime.now() - timedelta(days=60)).strftime('%Y-%m-%d')  # Lấy ngày hôm qua
to_date = (datetime.now() + timedelta(days=0)).strftime('%Y-%m-%d')

data = CommonBinanceDWH.CommonBinanceDWH.loaddataBinance_FromTo_Split_ListSymbol_Limit(symbol_list, from_date, to_date, '1d', 60)

print("Hoan thanh:")
print(datetime.now())
```

```python
data
```

```python
import talib	
data["MA20"] = talib.SMA(data["Close"], timeperiod=20)
data["MA50"] = talib.SMA(data["Close"], timeperiod=50)
data
```

```python
data["Buy_Signal"] = (data["MA20"] > data["MA50"]) 
data["Sell_Signal"] = (data["MA20"] < data["MA50"]) 
data
```
