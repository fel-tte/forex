# Lich su nap tien vao Binance

```python
import requests
import json
import matplotlib.pyplot as plt # External
from binance.client import Client # External
from datetime import datetime, timedelta # Built-in

import time
timestamp = int(time.time() * 1000)

API_KEY = api_key
API_SECRET = secret_key
client = Client(API_KEY, API_SECRET)

# Initialize variables
daily_amounts = {}
end_date = datetime.now()  # Ngày hiện tại
start_date = end_date - timedelta(days=15)  # 90 ngày trước

while start_date < end_date:
    # Tính startTime và endTime cho mỗi ngày
    startTime = int(start_date.timestamp()) * 1000
    endTime = int((start_date + timedelta(days=1)).timestamp()) * 1000

    # Lấy lịch sử nạp tiền
    deposits = client.get_deposit_history(startTime=startTime, endTime=endTime)    
    for deposit in deposits:
        date = datetime.fromtimestamp(deposit['insertTime'] / 1000).strftime('%Y-%m-%d')
        amount = float(deposit['amount'])
        daily_amounts[date] = daily_amounts.get(date, 0) + amount

    # Chuyển sang ngày tiếp theo
    start_date += timedelta(days=1)

dates = list(daily_amounts.keys())
amounts = list(daily_amounts.values())

plt.figure(figsize=(12, 6))
plt.plot(dates, amounts, marker='o', linestyle='-')
plt.title('Deposit History on Binance (Last 7200 Days)')
plt.xlabel('Date')
plt.ylabel('Amount (USDT)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Lich su nap tien vao Binance 2

```python
from binance.client import Client
from datetime import datetime, timedelta

# Khởi tạo client
client = Client(api_key, secret_key)

# Đặt khoảng thời gian bạn muốn lấy lịch sử nạp tiền, ví dụ: 60 ngày trước đây
end_date = datetime.now()
start_date = end_date - timedelta(days=90)

# Chuyển đổi ngày bắt đầu và kết thúc sang dấu thời gian tính bằng milliseconds
start_ts = int(start_date.timestamp() * 1000)
end_ts = int(end_date.timestamp() * 1000)

# Lấy lịch sử nạp tiền
deposits = client.get_deposit_history(startTime=start_ts, endTime=end_ts)
print(deposits)

# In kết quả
for deposit in deposits:
    timestamp = int(deposit['insertTime']) // 1000
    datedeposit = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d')
    print(datedeposit, deposit['coin'], deposit['amount'])
```
