# Ham history

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

def get_stock_data(ticker, period="1d", interval="1d"):
    """
    Lấy dữ liệu cổ phiếu từ Yahoo Finance
    
    Parameters:
    ticker (str): Mã cổ phiếu (ví dụ: 'AAPL' cho Apple)
    period (str): Khoảng thời gian ('1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max')
    interval (str): Khoảng thời gian giữa các điểm dữ liệu ('1m', '2m', '5m', '15m', '30m', '60m', '90m', '1h', '1d', '5d', '1wk', '1mo', '3mo')
    
    Returns:
    pandas.DataFrame: DataFrame chứa dữ liệu OHLC
    """
    # Tạo đối tượng Ticker
    stock = yf.Ticker(ticker)
    
    # Lấy dữ liệu lịch sử
    df = stock.history(period=period, interval=interval)
    
    return df
```

# Su dung ham download

```python
import yfinance as yf
import pandas as pd

# Lấy dữ liệu cổ phiếu ACB.VN
ticker = "ACB.VN"

# Sử dụng hàm download để lấy dữ liệu
df = yf.download(
    tickers=ticker,
    period="1d",  # Lấy dữ liệu 1 ngày
    interval="1d",  # Khoảng thời gian 1 ngày
)

# In thông tin dữ liệu
print(f"\nDữ liệu cổ phiếu {ticker}:")
print(df)

# Lưu dữ liệu vào file CSV
df.to_csv(f"{ticker}_data.csv")
print(f"\nĐã lưu dữ liệu vào file {ticker}_data.csv") 
```
