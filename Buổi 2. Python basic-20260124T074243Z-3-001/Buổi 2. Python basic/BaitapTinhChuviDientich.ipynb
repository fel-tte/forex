```python
import math

# Ham chu vi
def chuvi(r):
    chuvi = 2 * math.pi * r
    return chuvi

def dientich(r):
    dientich = math.pi * r * r 
    return dientich

# Nhập bán kính của đường tròn từ người dùng
r = float(input("Nhập bán kính của đường tròn: "))

chuvihinhtron = chuvi(r)
dientichhinhtron = dientich(r)

print(chuvihinhtron)
print(dientichhinhtron)
```
