# Dung ham

```python
import math
def chuvi(r):
	chuvi = 2 * math.pi * r
	return chuvi

def dientich(r):
	dientich = math.pi * r * r
	return dientich
```

```python
cv = chuvi(1)
cv
```

```python
dt = dientich(1)
dt
```

```python
# Buoc 1: Nhap
print("Moi nhap ban kinh duong tron")
# Buoc 1: Nhập bán kính của đường tròn từ người dùng
nhap = input("Nhập bán kính của đường tròn: ") # String
r = float(nhap) # Ep kieu cua nhap thanh float

# Buoc 2: Tính chu vi và diện tích bang cach goi ham
chu_vi = chuvi(r)
dien_tich = dientich(r)

# Buoc 3: Xuất kết quả
print(f"Chu vi của đường tròn là: {chu_vi}")
print('Chu vi của đường tròn là:', chu_vi)

print(f"Diện tích của đường tròn là: {dien_tich}")
```

```python
import math

def dientichExt(r):
	return math.pi * r**2

# Nhập bán kính của đường tròn từ người dùng
try:
	r = float(input("Nhập bán kính của đường tròn: "))
	
	if r <= 0:
		raise ValueError("Bán kính phải là một số dương khác không.")
except ValueError as e:
	print(f"Lỗi: {e}")
	exit()

# Tính chu vi và diện tích
# chu_vi = 2 * math.pi * r
chu_vi = chuvi(r)
# dien_tich = math.pi * r**2
dien_tich = dientichExt(r)

# Xuất kết quả
print(f"Chu vi của đường tròn là: {chu_vi}")
print(f"Diện tích của đường tròn là: {dien_tich}")
```
