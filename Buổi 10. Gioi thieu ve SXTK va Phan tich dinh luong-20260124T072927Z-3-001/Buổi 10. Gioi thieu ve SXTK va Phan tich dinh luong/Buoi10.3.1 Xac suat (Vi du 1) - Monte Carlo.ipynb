# Mo phong Monte Carlo

```python
import random
import matplotlib.pyplot as plt

# Thông số mô phỏng
num_simulations = 1000  # Số lần chạy mô phỏng
num_trials = 1000  # Số lần tung đồng xu trong mỗi mô phỏng

# Lưu trữ kết quả xác suất của mỗi lần mô phỏng
probabilities = []

for _ in range(num_simulations):
    results = [random.choice(['S', 'N']) for _ in range(num_trials)]
    prob_S = results.count('S') / num_trials
    probabilities.append(prob_S)

# Vẽ biểu đồ phân phối xác suất
plt.hist(probabilities, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(x=0.5, color='r', linestyle='dashed', linewidth=2, label="Xác suất lý thuyết (0.5)")
plt.xlabel("Xác suất xuất hiện Sấp (S)")
plt.ylabel("Số lần xuất hiện")
plt.title(f"Phân phối xác suất sau {num_simulations} lần mô phỏng Monte Carlo")
plt.legend()
plt.show()
```
