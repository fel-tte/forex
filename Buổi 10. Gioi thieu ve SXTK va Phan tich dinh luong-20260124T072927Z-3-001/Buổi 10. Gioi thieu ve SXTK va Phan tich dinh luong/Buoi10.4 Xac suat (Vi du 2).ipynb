##### Để mô phỏng việc gieo một đồng xu bất cân đối 10,000 lần, bạn cần xác định trọng lượng cho mỗi mặt của đồng xu. 
##### Ví dụ, giả sử trọng lượng để đồng xu xuất hiện mặt Sấp là 60% và Ngửa là 40%. 
##### Bạn có thể sử dụng thư viện random trong Python để mô phỏng điều này.

```python
import random

# Định nghĩa số lần gieo đồng xu
num_throws = 10000

# Trong luong mặt Sấp
tl_s = 60/100
tl_n = 40/100

# Tạo một list để lưu trữ kết quả của mỗi lần gieo
outcomes = []

# Gieo đồng xu và lưu kết quả vào list
for _ in range(1, num_throws + 1, 1): # _ là một biến không được sử dụng, range(num_throws) tạo ra một dãy số từ 0 đến num_throws-1
    if random.random() <= tl_s: # random.random() tạo ra một số ngẫu nhiên từ 0 đến 1
        outcomes.append('S')  # Mặt Sấp
    else:
        outcomes.append('N')  # Mặt Ngửa

# Đếm số lần xuất hiện của mỗi kết quả
s_count = outcomes.count('S')
n_count = outcomes.count('N')

# In kết quả
print(f"Số lần Sấp: {s_count}")
print(f"Số lần Ngửa: {n_count}")
print(f"Phan tram Sấp: {s_count/num_throws*100}")
print(f"Phan tram Ngửa: {n_count/num_throws*100}")
```
