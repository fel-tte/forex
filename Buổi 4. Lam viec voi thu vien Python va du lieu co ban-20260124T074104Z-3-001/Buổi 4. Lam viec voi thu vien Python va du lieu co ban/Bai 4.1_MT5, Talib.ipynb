# Bai tap: DA lay duoc data MT5 : Dang lay luong tuong
=> Viet ham de lay data: truyen vao symbol va timeframe

```python
import MetaTrader5 as mt5
import pandas as pd

def getDataMT5(symbol, timeframe, n):
	"""
	Lấy dữ liệu giá từ MetaTrader 5
	:param symbol: Tên cặp tiền tệ
	:param timeframe: Khung thời gian (ví dụ: mt5.TIMEFRAME_M30)
	:param n: Số lượng nến cần lấy
	:return: DataFrame chứa dữ liệu giá
	"""

	rates = mt5.copy_rates_from_pos(symbol, timeframe, 0, 1000) # Method lấy dữ liệu từ MetaTrader 5 da co o trên
	print(rates)
	dataFrame = pd.DataFrame(rates)
	print(dataFrame)

	dataFrame['time'] = pd.to_datetime(dataFrame['time'], unit='s')
	dataFrame.set_index('time', inplace=True)
	
	dataFrame = dataFrame.rename(columns={
		'open': 'Open',
		'high': 'High',
		'low': 'Low',
		'close': 'Close',
		'tick_volume': 'Volume'
	})
	
	return dataFrame[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
# Goi ham
data = getDataMT5("EURUSD.sml", mt5.TIMEFRAME_D1, 1000)
data
```
