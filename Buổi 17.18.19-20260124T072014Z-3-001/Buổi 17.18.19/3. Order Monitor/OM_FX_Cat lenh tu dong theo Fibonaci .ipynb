```python
# Hàm tính toán mức Fibonacci
def calculate_fibonacci_levels(price_high, price_low):
    level_61_8 = price_high - 0.618 * (price_high - price_low)
    level_38_2 = price_high - 0.382 * (price_high - price_low)
    level_50 = (price_high + price_low) / 2
    return level_38_2, level_50, level_61_8
```

```python
import MetaTrader5 as mt5
def closeOrder(symbol):
    # Lay tat ca cac lenh dang mo
    positions = mt5.positions_get(symbol=symbol)

    if positions is None or len(positions) == 0:
        print(symbol, "Không có vị thế nào đang mở")
    else:
        for position in positions:
            close_result = mt5.Close(symbol, ticket=position.ticket)
            # mt5.order_close(ticket, volume, price, deviation)
            # mt5.close_position    
            
            # Kiểm tra kết quả đóng lệnh
            if close_result:
                print(symbol, 'Đã đóng lệnh thành công rồi. Ticket:', position.ticket)
            else:
                print(symbol, "Lệnh đóng thất bại, error code =", mt5.last_error())     
```

```python
import MetaTrader5 as mt5

# Giám sát lệnh và xác định điểm thoát
def monitor_orders(symbol, fib_levels):
    # Giả định có giá thị trường hiện tại
    current_price = mt5.symbol_info_tick(symbol).ask
    # So sánh giá hiện tại với mức Fibonacci để quyết định điểm thoát
    if current_price >= fib_levels[0]:
        print(symbol, "Đã đạt đến mức 38.2%, cân nhắc thoát lệnh.")
        closeOrder(symbol)
    elif current_price >= fib_levels[1]:
        print(symbol, "Đã đạt đến mức 50%, cân nhắc thoát lệnh.")
        closeOrder(symbol)
    elif current_price >= fib_levels[2]:
        print(symbol, "Đã đạt đến mức 61.8%, cân nhắc thoát lệnh.")
        closeOrder(symbol)
    else:
        print(symbol, "Giám sát tiếp tục. Mọi thứ đang ổn")
```
