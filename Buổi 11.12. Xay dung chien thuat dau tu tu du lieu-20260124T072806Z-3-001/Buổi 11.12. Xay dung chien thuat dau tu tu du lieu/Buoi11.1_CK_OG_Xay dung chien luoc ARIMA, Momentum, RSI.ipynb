# Processing data, Signal

```python
import talib as ta
from datetime import datetime, timedelta
import statsmodels.api as sm

def scan_Signal(symbol, from_date, to_date, timeframe):

	data = TradeYFinance.TradeYFinance.loaddataYFinance(symbol, from_date, to_date, timeframe)
	data['RSI'] = ta.RSI(data['Close'], timeperiod=14).dropna() # 14 days
	data['Momentum'] = ta.MOM(data['Close'], timeperiod=10).dropna() # 10 days

	# Dự đoán giá, them 1 cot moi vao DataFrame
	model = sm.tsa.ARIMA(data['Close'], order=(3, 5, 0)) # Chay toi uu o trên, o day chi can dua vao ket qua da chay o tren
	model_fit = model.fit()

	data['Predicted_Close'] = model_fit.predict(start=0, end=len(data)-1, typ='levels') #$ Dự đoán giá từ đầu đến cuối của chuỗi thời gian

	# Xác định tín hiệu mua và bán
	# Tin hieu mua xay ra Gia du doan > Close va M > 0 va R < 30
	data['Buy_Signal'] = ((data['Predicted_Close'] > data['Close']) & (data['Momentum'] >= 0) & (data['RSI'] <= 30))
	data['Sell_Signal'] = ((data['Predicted_Close'] < data['Close']) & (data['Momentum'] < 0) & (data['RSI'] >= 70))
	
	# Day sang redis voi last_record
	# Tạo kết nối Redis
	import redis
	import pandas as pd
	import numpy as np

	r = redis.Redis(host='localhost', port=6379, db=15) # 0-5 CK, 6-10 Crypto, 11-15 Forex
	# Tạo hash key
	hash_key = 'OG_FX_ML, Momentum, RSI_K13'
	last_record = data.iloc[-1]
   
	# Chuyển đổi record cuối cùng thành từ điển và lưu vào Redis dưới dạng hash
	if(last_record['Buy_Signal'] == True or last_record['Sell_Signal'] == True):
		for field, value in last_record.to_dict().items():
			# Chuyển đổi giá trị uint64 và Timestamp thành chuỗi
			if isinstance(value, pd.Timestamp):
				value = value.isoformat()
			elif isinstance(value, (int, np.uint64)):
				value = str(value)
			r.hset(hash_key, field, value)  
			r.hset(hash_key, 'Symbol', symbol)
			r.hset(hash_key, 'Insertdate', datetime.now().isoformat())
			last_record['Insertdate'] = datetime.now().isoformat()
		print(last_record)   
	else:
		print(last_record)   
		print('Không có tín hiệu!')
```

```python
# 1 ngay chay 1 lan tai thoi diem 10h30
import schedule
import time
from datetime import datetime, timedelta

def schedule_Signal():
	print("Chay tai thoi diem", datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
	symbol = 'VCB.VN'
	# from_date = '2024-11-01'
	# to_date = '2024-10-16' # yyyy-mm-dd # to_date = to - 1
	from_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')  # Lấy ngày hôm qua
	to_date = (datetime.now() + timedelta(days=1)).strftime('%Y-%m-%d')  # Lấy ngày hôm qua
	timeframe = '1d'
	scan_Signal(symbol, from_date, to_date, timeframe)

schedule.every().day.at("21:17").do(schedule_Signal)

while True:
	schedule.run_pending()
	# time.sleep(1)
```
