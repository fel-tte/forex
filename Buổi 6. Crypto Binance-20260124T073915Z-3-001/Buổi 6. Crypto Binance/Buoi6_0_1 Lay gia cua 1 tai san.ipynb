# Binance

ccxt

```python
import ccxt
import pandas as pd
from datetime import datetime

# Initialize the Binance exchange
exchange = ccxt.binance()

# Define the symbol for Bitcoin traded against USDT on Binance
symbol = 'BTC/USDT'

# Initialize an empty list to store historical prices
historical_prices_list = []

# Loop through each year from the earliest available data on Binance to 2023
for year in range(2017, 2024):  # Binance started in mid-2017
    # Construct the timestamp for December 31st of the current year
    date_str = f'{year}-12-31T00:00:00Z'
    # Convert the date string to a timestamp (milliseconds)
    since = exchange.parse8601(date_str)
    
    # Fetch historical OHLCV data from Binance
    ohlcv = exchange.fetch_ohlcv(symbol, '1d', since, limit=1)
    
    # Check if data was fetched
    if ohlcv:
        # Extract the closing price
        close_price = ohlcv[0][4]
        # Append the data to the list
        historical_prices_list.append({'Date': date_str[:10], 'Closing Price (USDT)': close_price})
    else:
        # Handle case where no data was returned
        historical_prices_list.append({'Date': date_str[:10], 'Closing Price (USDT)': 'None'})

# Convert the list to a DataFrame
historical_prices_df = pd.DataFrame(historical_prices_list)

# Display the historical prices DataFrame
print(historical_prices_df)
```

binance client

```python
from binance.client import Client
import pandas as pd
import dateutil.parser

# Initialize the Binance client without API key and secret (for public data)
client = Client()

def fetch_binance_btc_price_on_date(date_str):
    # Convert the date to milliseconds
    date_ms = int(dateutil.parser.parse(date_str).timestamp() * 1000)
    # Fetch the historical klines (OHLCV) data
    klines = client.get_historical_klines("BTCUSDT", Client.KLINE_INTERVAL_1DAY, date_ms, date_ms + (24 * 60 * 60 * 1000))
    
    # Check if any data was returned
    if klines:
        # Extract the closing price from the first (and only) kline
        closing_price = klines[0][4]
        return closing_price
    else:
        return None

# Define a list of dates (December 31st from 2017 to 2023)
dates = [f"{year}-12-31" for year in range(2009, 2024)]

# Fetch the prices
prices = [fetch_binance_btc_price_on_date(date) for date in dates]

# Create a DataFrame to display the results
df_prices = pd.DataFrame(list(zip(dates, prices)), columns=['Date', 'Closing Price (USDT)'])

print(df_prices)
```

# YFinance

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for Bitcoin on Yahoo Finance
ticker_symbol = 'BTC-USD'

# Initialize an empty list to store the results
historical_prices = []

# Define the range of years where Bitcoin data is more likely to be available
start_year = 2009  # Adjusted based on your output
end_year = 2023

for year in range(start_year, end_year + 1):
    # Construct the date strings
    start_date = f"{year}-12-31"
    end_date = f"{year+1}-01-01"
    
    # Fetch data from Yahoo Finance
    data = yf.download(ticker_symbol, start=start_date, end=end_date)
    
    # Check if any data was fetched
    if not data.empty:
        # Correctly use .iloc to avoid FutureWarning
        close_price = data['Close'].iloc[0]
        # Append the date and closing price to the list
        historical_prices.append((start_date, close_price))

# Convert the list to a DataFrame
historical_prices_df = pd.DataFrame(historical_prices, columns=['Date', 'Close'])

# Display the historical prices DataFrame
print(historical_prices_df)
```
