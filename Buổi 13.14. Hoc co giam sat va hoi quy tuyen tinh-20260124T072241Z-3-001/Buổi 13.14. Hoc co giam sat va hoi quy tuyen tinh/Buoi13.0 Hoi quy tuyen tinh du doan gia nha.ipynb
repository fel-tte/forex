Xây dựng một mô hình học máy dự đoán giá nhà dựa trên diện tích đất (tính bằng feet vuông). Đây là một ví dụ điển hình của bài toán hồi quy, trong đó biến đầu vào (tính năng) là diện tích nhà và biến đầu ra (mục tiêu) là giá của nhà. Một mô hình hồi quy tuyến tính đơn giản có thể được sử dụng để hiểu và dự đoán mối quan hệ này, vì giá nhà thường tăng theo diện tích.

# Biểu diễn các giá trị trong bảng trên dưới dạng một biểu đồ phân tán, với các giá trị được hiển thị bởi các điểm đánh dấu màu đỏ

Nhiem vu la tim ra duong hoi quy

<img src="figs/scatterplot.jpeg" style="width:600px;height:370px">

```python
import matplotlib.pyplot as plt
import numpy as np

# Các giá trị cho 'area' (được lấy từ trục hoành của scatter plot)
area = np.array([2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000])

# Áp dụng phương trình hồi quy để tính 'price'
price = 135.78767123 * area + 180616.43835616432

# Tạo scatter plot với dữ liệu hiện có
plt.scatter(area, price, color='red')  # Các điểm dữ liệu

# Tạo và vẽ đường hồi quy
plt.plot(area, price, color='blue', linewidth=3)  # Đường hồi quy

# Đặt nhãn cho trục x và y
plt.xlabel('Area')
plt.ylabel('Price')

# Hiển thị biểu đồ
plt.show()
```

Quá trình tối ưu hóa một mô hình hồi quy tuyến tính, nơi chúng ta cố gắng tìm ra đường thẳng phù hợp nhất với các điểm dữ liệu. Mục tiêu là để "tổng bình phương sai số" (tổng bình phương của các khoảng cách từ từng điểm dữ liệu đến đường hồi quy) là nhỏ nhất. Các khoảng cách này, thường được gọi là "sai số" hoặc "residuals" (A1, A2 trên hình), thể hiện sai số giữa giá trị thực tế và giá trị dự đoán bởi mô hình. Việc tối thiểu hóa tổng bình phương của các sai số này giúp đảm bảo rằng mô hình là tốt nhất có thể, phản ánh đúng mối quan hệ giữa biến độc lập và biến phụ thuộc trong dữ liệu.

```python
import pandas as pd # Cài đặt thư viện cần thiết: Load dữ liệu và xử lý
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model # Hồi quy tuyến tính
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='.')
```

# Lay bien doc lap

```python
X = df.drop('price', axis='columns') # Xóa cột 'price' để chỉ giữ lại cột 'area'
X
```

# Lay bien phu thuoc

```python
Y = df.price
Y
```

# Khoi tao

```python
# Tạo đối tượng hồi quy tuyến tính
model = linear_model.LinearRegression() # Tạo mô hình hồi quy tuyến tính

model.fit(X, Y) # Huấn luyện mô hình với dữ liệu
```

```python
df
```

**(1) Dự đoán giá nhà với diện tích = 5000**

```python
# Chung ta co mo hinh model
# In hệ số góc và hệ số chặn của đường hồi quy
model.predict([[8000]])  # Dự đoán giá cho diện tích 5000
```

# Tinh ra he so goc

```python
model.coef_ # In hệ số góc (slope): m
```

# Tinh ra he so chan

```python
model.intercept_ # In hệ số chặn (intercept): b
```

# The vao ham so y = mx +

```python
model.coef_ * 8000 + model.intercept_
```

# The vao duong hoi quy

```python
# O tren: du doan bang model.predict([[5000]])
model.predict([[5000]])  # Dự đoán giá cho diện tích 5000
# O duoi: phuong trinh du doan bang model.coef_ va model.intercept_
```

```python
# The vao cong thuc y = mx + b
model.coef_[0] * 5000 + model.intercept_  # Tính giá dự đoán cho diện tích 5000
```

```python
model.predict([[5000]])
```

**Y = m * X + b (m la he so goc va b la he so chan)**
