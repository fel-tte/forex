# Buoc 1: Load data

```python
import sys 
sys.path.append("../Common")

import CommonMT5 
```

```python
from datetime import datetime, timedelta
import MetaTrader5 as mt5

symbol = "EURUSD.sml"
from_date = (datetime.now() - timedelta(days=0)).strftime("%Y-%m-%d")
to_date = (datetime.now() + timedelta(days=2)).strftime("%Y-%m-%d")
timeframe = mt5.TIMEFRAME_M30

data = CommonMT5.CommonMT5.loaddataMT5_FromTo(symbol, from_date, to_date, timeframe)
data
```

# Buoc 2: Xu ly data

```python
# Gia du doan dua vao hoi quy tuyen tinh
# Du doan gia dong cua dua tren gia dong dua vao gia dong cua ngay hom truoc
import pandas as pd
import numpy as np
import sklearn

model = sklearn.linear_model.LinearRegression()

X = data['Close'].shift(1).fillna(0)
X = pd.DataFrame(X)

print(len(X))

Y = data['Close'].fillna(0)
Y = pd.DataFrame(Y)
print(len(Y))

model.fit(X, Y)

data['Predict_Close'] = model.predict(pd.DataFrame(data['Close']))
```
