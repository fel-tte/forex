```python
import pandas as pd

dataFrame = pd.DataFrame(thisdict)
print(dataFrame)
dataFrame.to_csv('dataFrame.csv')
dataFrame.to_excel('dataFrame.xlsx')
dataFrame.to_json('dataFrame.json')
dataFrame.to_html('dataFrame.html')
```

```python
from fpdf import FPDF
from tabulate import tabulate

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=10)

table = tabulate(dataFrame.values, headers=dataFrame.columns, tablefmt="grid")
for row in table.split('\n'):
    pdf.cell(0, 5, row, ln=1)
pdf.output("dataFrame.pdf")
```

```python
import pandas as pd
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors

# Giả sử bạn đã có DataFrame tên là dataFrame
dataFrame = pd.DataFrame({
    'Tên': ['A', 'B', 'C'],
    'Tuổi': [23, 34, 45],
    'Điểm': [8.5, 7.0, 9.0]
})

pdf_file = "dataFrame1.pdf"
doc = SimpleDocTemplate(pdf_file, pagesize=landscape(letter))

# Chuyển DataFrame thành list cho Table
data = [dataFrame.columns.to_list()] + dataFrame.values.tolist()

# Tạo bảng
table = Table(data)
style = TableStyle([
    ('BACKGROUND', (0,0), (-1,0), colors.grey),
    ('TEXTCOLOR',(0,0),(-1,0),colors.whitesmoke),
    ('ALIGN',(0,0),(-1,-1),'CENTER'),
    ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
    ('BOTTOMPADDING', (0,0), (-1,0), 12),
    ('BACKGROUND',(0,1),(-1,-1),colors.beige),
    ('GRID', (0,0), (-1,-1), 1, colors.black),
])
table.setStyle(style)

elements = [table]
doc.build(elements)
print("Đã xuất ra file PDF đẹp:", pdf_file)
```
