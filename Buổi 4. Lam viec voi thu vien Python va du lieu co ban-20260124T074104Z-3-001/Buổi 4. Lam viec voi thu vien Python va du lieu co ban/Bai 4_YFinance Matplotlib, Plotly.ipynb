```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = 'VCB.VN'

# Get data on this ticker
ticker_data = yf.Ticker(ticker_symbol)

# Get the historical prices for this ticker
ticker_df = ticker_data.history(period='1mo', interval='1m', start='2025-09-15', end='2025-09-20')

# Display the data
# ticker_df.head()
ticker_df
```

# Bai tap: Viet ham de lay du lieu cua ma chung khoan ACB tren Yfinance trong vong 10 ngay, voi interval = 5m

```python
import yfinance as yf
def getDataYFinance(ticker, start, end, interval='5m'):
	"""
	Fetch historical data for a given ticker symbol from Yahoo Finance.
	
	Parameters:
	ticker (str): The ticker symbol to fetch data for.
	start (str): The start date in 'YYYY-MM-DD' format.
	end (str): The end date in 'YYYY-MM-DD' format.
	interval (str): The data interval (default is '5m').
	
	Returns:
	DataFrame: A pandas DataFrame containing the historical data.
	"""
	return yf.download(ticker, start=start, end=end, interval=interval)	

# Goi ham
data = getDataYFinance('AAPL', '2025-09-01', '2025-09-25', interval='5m')
data
```
