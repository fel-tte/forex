# Co 2 thu vien: ccxt, binance

```python
# Xai ccxt
# Dung ccxt de lay du lieu tu Binance
import ccxt
import pandas as pd
from datetime import datetime

# Khởi tạo exchange Binance
exchange = ccxt.binance() # ccxt co the ket noi den hon 150 san Crypto toan the gioi

# Lấy dữ liệu OHLCV (Open, High, Low, Close, Volume)
symbol = 'BTC/USDT'  # Cặp tiền
timeframe = '1h'     # Khung thời gian: 1m, 5m, 15m, 1h, 4h, 1d, 1w, 1M

# Lấy 100 nến gần nhất
ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=100)

# Chuyển thành DataFrame
df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])

# Chuyển timestamp thành datetime
df['datetime'] = pd.to_datetime(df['timestamp'], unit='ms') # Timestamp dang miligiay
df.set_index('datetime', inplace=True)

print(df.tail())
```

```python
def getData_Binance(symbol, timeframe, limit=100):
	# Xai ccxt
	# Dung ccxt de lay du lieu tu Binance
	import ccxt
	import pandas as pd
	from datetime import datetime

	# Khởi tạo exchange Binance
	exchange = ccxt.binance() # ccxt co the ket noi den hon 150 san Crypto toan the gioi

	# Lấy dữ liệu OHLCV (Open, High, Low, Close, Volume)
	symbol = 'BTC/USDT'  # Cặp tiền
	timeframe = '1h'     # Khung thời gian: 1m, 5m, 15m, 1h, 4h, 1d, 1w, 1M

	# Lấy 100 nến gần nhất
	ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=100)

	# Chuyển thành DataFrame
	df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])

	# Chuyển timestamp thành datetime
	df['datetime'] = pd.to_datetime(df['timestamp'], unit='ms') # Timestamp dang miligiay
	df.set_index('datetime', inplace=True)

	return df
```
